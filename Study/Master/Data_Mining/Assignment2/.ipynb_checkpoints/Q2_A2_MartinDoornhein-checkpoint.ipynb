## Question 2
BN'er is an abbreviation in the Dutch language. This abbreviation means something like famous Dutch person or celebrity. The query ‘BN’er’ yields many results on google of faces that are known to many Dutch people from television and social media. Many of the photos of these celebrities are distributed by social media and news sites. Celebrities have little influence on what is published about them on the internet. That is why it seems interesting what image media creates with the publication of photos of these people. To investigate this, we look at which emotion can be read on the face of the celebrities. 
According to Ekman (1999) there are six basic emotions which are universal for human. These emotions are: angry, disgust, fear, happy, sad, surprise. This research will use this list of emotions and add a seventh which is: neutral. 
These six emotions can be divided in positive and negative emotions. Angry, happy, fear and sad will be classified as negative. Happy and surprise will be classified as positive.
So the research question is: Does media image our Dutch celebrities with positive or negative emotions.  
We will scrape 50 images from google with the python library: Google Image download. To investigate the emotions in the picture, we will use a face recognition algorithm which is capable of detection emotions. We will use the probability of the highest emotion in the image, because there could be multiple people in one image. The result of this analysis gives a conclusion on what percentage of the images could be classified as positive vs. negative.


*Reference*  
Ekman, P. (1999). Basic emotions. Handbook of cognition and emotion, 98(45-60), 16.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
from tqdm.notebook import tqdm

from PIL import Image, ImageOps
from tensorflow.keras.preprocessing import image

import wget
import cv2

from tensorflow.keras.models import load_model

%config InlineBackend.figure_format='retina' 
```

```python
#scraping images from google with google_image_download
#code from stackoverflow

from google_images_download import google_images_download
response = google_images_download.googleimagesdownload()

arguments = {"keywords":"BN'er","limit":50,"print_urls":True}
paths = response.download(arguments)
```

```python
#creating a dataframe with for loop
#code frome stackoverflow: 
#https://stackoverflow.com/questions/10377998/how-can-i-iterate-over-files-in-a-given-directory
import os

directory = os.fsencode("downloads/BN'er/")

images_paths = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    file_path = os.fsdecode(directory+file)
    image_df = {'Image_title': filename,
               'Image_path': file_path}
    images_paths.append(image_df)

df = pd.DataFrame(images_paths)
```

```python
df.head()
```

```python
#loading images from path
#code from exercise 5.2
def load_image_from_path(image_path, target_size=None, color_mode='rgb'):
    pil_image = image.load_img(image_path, 
                               target_size=target_size,
                            color_mode=color_mode)
    return image.img_to_array(pil_image)

#imagepath
image_path = df["Image_path"][4]
color_image = load_image_from_path(image_path)

#example picture -- Ali B
plt.imshow(color_image.astype(np.uint8))
```

```python
#url to download our model from
model_url = 'https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml'
face_model = wget.download(model_url)

#face classifcation model
face_classification = cv2.CascadeClassifier(face_model)
```

```python
def apply_offsets(face_coordinates, offsets):
    """
    Derived from https://github.com/oarriaga/face_classification/blob/
    b861d21b0e76ca5514cdeb5b56a689b7318584f4/src/utils/inference.py#L21
    """
    x, y, width, height = face_coordinates
    x_off, y_off = offsets
    return (x - x_off, x + width + x_off, y - y_off, y + height + y_off)

#Emtotion recognition url to download our model
model_url = "https://github.com/oarriaga/face_classification/raw/master/trained_models/emotion_models/fer2013_mini_XCEPTION.102-0.66.hdf5"
emotion_model = wget.download(model_url)

#loading face classification model
emotion_classifier = load_model('fer2013_mini_XCEPTION.102-0.66.hdf5')
EMOTION_OFFSETS = (0, 0)
INPUT_SHAPE_EMOTION = emotion_classifier.input_shape[1:3]

#emotions categories
labels = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

```

```python
#for loop for identifying the number of faces and the emotions

#list for dataframe
emotion_list = []
n_faces = []


for file in tqdm(df.Image_path):

    #load image 
    pre_image = load_image_from_path(file, color_mode='grayscale')
    gray_image = np.squeeze(pre_image).astype('uint8')
  

    #face counter
    faces = face_classification.detectMultiScale(gray_image, 1.3, 5)
    n_faces.append(len(faces))
    
    #emotion recognition
    for face_coordinates in faces:
        x1, x2, y1, y2 = apply_offsets(face_coordinates, EMOTION_OFFSETS) 
        face_img = gray_image[y1:y2, x1:x2] # only get the face
        face_img = cv2.resize(face_img, (INPUT_SHAPE_EMOTION))
        face_img = face_img.astype('float32') / 255.0 # pre-processing 
        face_img = face_img - 0.5 # pre-processing specific to the emotion classifier
        face_img = face_img * 2.0 # pre-processing specific to the emotion classifier
        face_img = np.expand_dims(face_img, 0) # batch of one
        face_img = np.expand_dims(face_img, -1) # pre-processing specific to the emotion classifier
        probas = emotion_classifier.predict(face_img)
        label = labels[np.argmax(probas[0])]
        emotionprobas = dict(zip(labels, probas[0])) # make a dictionary from the labels and the probabilities
        
        #get highest emotion probabilty
        #from stackoverflow
        maximum = max(emotionprobas, key=emotionprobas.get)
    #get highest probabilty (in case there are multiple faces/emotions)
    emotion_list.append(maximum)

#add resutls to df
df['n_faces'] = n_faces
df['emotion'] = emotion_list
```

```python
ax = df['emotion'].value_counts().plot(kind = 'barh',
                                          figsize = (16,9),
                                          fontsize = 18)
                                          
ax.set_title('Absolute frequency of emotions', fontsize = 18)
ax.set_xlabel('Absolute frequency', fontsize = 18)
ax.set_ylabel('Emotions', fontsize = 18)
```

```python
#catgorizing in positive, negative en neutral emotions
df.loc[(df["emotion"] == 'happy') | (df["emotion"] == 'surprise'), 'emotion_cat'] = 'positive'
df.loc[(df["emotion"] == 'angry') | (df["emotion"] == 'fear') | (df["emotion"] == 'sad'), 'emotion_cat'] = 'negative'
df.loc[(df["emotion"] == 'neutral'), 'emotion_cat'] = 'neutral'

#value_counts aboslute and relative
print(df.emotion_cat.value_counts(), "\n")
print(df.emotion_cat.value_counts(normalize=True))
```

```python
#barplot aboslute frequencies
ax = df['emotion_cat'].value_counts().plot(kind = 'barh',
                                          figsize = (16,9),
                                          fontsize = 18)
                                          
ax.set_title('Absolute frequency of emotions', fontsize = 18)
ax.set_xlabel('Absolute frequency', fontsize = 18)
ax.set_ylabel('Emotions', fontsize = 18)
```

## Interpretation and conlcusion

According to this analysis the 50 images of BN'ers scraped from Google, are for the biggest part classified as positive emotions (54%). The negative emotion category consisted of 26% and the neutral category consisted of 19%. According to this result we can say that based on this small amount of images, it looks like media chooses to image BN'ers more frequently with positive emotions than negative or neutral emotions. A remarkable finding is that we didn’t classify any image with the emotion disgust.

But, we should interpret these results very carefully. After manual inspection of some rows of our dataframe in comparison to the actual images, we can say that the face detection algorithm is not so accurate. It sometimes not recognizes multiple faces and also sometimes doesn’t recognize any faces at all. 
To really dive into a research question like this one, there are some things which should be done differently. In the first place, 50 images is way too small dataset. Also there should be images from different sources if we want to get insight in how media does position BN’ers with images.
