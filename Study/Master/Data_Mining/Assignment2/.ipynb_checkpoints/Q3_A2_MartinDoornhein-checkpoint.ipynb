### It’s a Man’s World
According to a recent study, female protagonists in top movies increased the last few years (Lauzen, 2019). In 2016 the percentage of woman protagonists was 16% which increased to 40% in 2019. This research will look at movie trailers (dataset: movietrailers.csv) from different time periods and tries to detect the number of females in it. We will try to find the proportion of females vs. males actors per subset in the movie trailer.  To select the movie trailers, we will use three subsets from the movietrailer.csv dataset and will sample 8 trailers from each subset. The subsets will be made based on the year.    
* Subset 1: 1920-1940   
* Subset 2: 1960-1980   
* Subset 3: 2000-2020  

Each movie trailer will be split in scenes which will be done with the scenedetect package. After this split, the algorithm will choose the frame exactly in the middle and adds this to our list which we will analyze. We choose this method because it drastically reduces the number of frames, which is beneficial for our computational load. Also we exepect that using every frame for example, will lead to a lot of frames which are useless.

To deal with the dimensionality of the data, we use the average proportion of males vs female per time period. So, there will be three dataframes which contains the selected frames for the given time period. Every dataframe has a column for number of detected faces (n_faces), number of detected females (n_females) and number of detected males (n_males). These datasets give insight in how the individual trailers contribute to the final dataframe average.

Comparing these dataframes should lead to the answer of our research question:    
**Does the proportion of females in movies increased over time?**

*Reference:*   
Lauzen, M. M. (2019). It’sa Man’s (Celluloid) World: Portrayals of Female Characters in the Top Grossing Films of 2018. Center for the Study of Women in Television and Film, San Diego State University.

```python
#function for scenedetection 
#code from manual

def find_scenes(video_path, threshold=30.0):
    video_manager = VideoManager([video_path])
    scene_manager = SceneManager()
    scene_manager.add_detector(
        ContentDetector(threshold=threshold))
    base_timecode = video_manager.get_base_timecode()
    video_manager.set_downscale_factor()
    video_manager.start()
    scene_manager.detect_scenes(frame_source=video_manager, show_progress=True)    
    return scene_manager.get_scene_list(base_timecode)
```

```python
#function for downloading the frames
#code from manual
def get_scene_frames(df_paths):
    for path in df_paths:
        scene_list = find_scenes(path, threshold=30)
        cap = cv2.VideoCapture(path)

        frames = []

        for start_time, end_time in scene_list:
            duration = end_time - start_time
            frame = (start_time.get_frames() + int(duration.get_frames() / 2))
            cap.set(cv2.CAP_PROP_POS_FRAMES,frame)
            ret, frame = cap.read()
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frames.append(frame)
        filename = path.split("/",3)[-1]


        #save frames
        if not os.path.exists('scenes/'+ filename +'/'):
            os.mkdir('scenes/'+ filename + '/')

        for i, frame in enumerate(frames):
            cv2.imwrite('scenes/'+filename+'/frame_{}.jpg'.format(i), frame)
```

```python
#loading images from path
#code from exercise 5.2
def load_image_from_path(image_path, target_size=None, color_mode='rgb'):
    pil_image = image.load_img(image_path, 
                               target_size=target_size,
                            color_mode=color_mode)
    return image.img_to_array(pil_image)
```

```python
#code from exercise 5.2

#url to download our model from
model_url = 'https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml'
face_model = wget.download(model_url)

#face classifcation model
face_classification = cv2.CascadeClassifier(face_model)

model_url = "https://github.com/oarriaga/face_classification/raw/master/trained_models/gender_models/gender_mini_XCEPTION.21-0.95.hdf5"
gender_model = wget.download(model_url)

def apply_offsets(face_coordinates, offsets):
    """
    Derived from https://github.com/oarriaga/face_classification/blob/
    b861d21b0e76ca5514cdeb5b56a689b7318584f4/src/utils/inference.py#L21
    """
    x, y, width, height = face_coordinates
    x_off, y_off = offsets
    return (x - x_off, x + width + x_off, y - y_off, y + height + y_off)

gender_classifier = load_model('gender_mini_XCEPTION.21-0.95.hdf5') # load this only once! (not in a loop)

GENDER_OFFSETS = (10, 10)
INPUT_SHAPE_GENDER = gender_classifier.input_shape[1:3]

labels = ['woman', 'man']
```

#### Note
To classify the gender of the people in the frames, we have first detects faces. 
After face detection we can try to classify their gender.
For some reason, this code below didn't give the resutls we wanted in the format of fucntion.
So, the loop for detecting faces and classifying gender is made 3 times for the three time periods.

```python
#loop for df_frames_1
#code from manual

#list for dataframe
n_faces = []
n_male = []
n_female = []

#loop to detect gender from frames
for file in tqdm(df_frames_1.path):

    #load image 
    pre_image = load_image_from_path(file, color_mode='grayscale')
    gray_image = np.squeeze(pre_image).astype('uint8')  

    #face counter
    faces = face_classification.detectMultiScale(gray_image, 1.3, 5)
    n_faces.append(len(faces))

    male = 0
    female = 0

    for face_coordinates in faces: # using the output of the CascadeClassifier
        try:
            x1, x2, y1, y2 = apply_offsets(face_coordinates, GENDER_OFFSETS) # extends the bounding box
            face_img = gray_image[y1:y2, x1:x2] # only get the face 
            face_img = cv2.resize(face_img, (INPUT_SHAPE_GENDER)) # resize the image
            face_img = face_img.astype('float32') / 255.0 # preprocess the image
            face_img = np.expand_dims(face_img, 0) # batch of one
            probas = gender_classifier.predict(face_img)
        except:
            print('Error..continuing')
            continue

        if labels[np.argmax(probas[0])] == 'man':
            male += 1
        elif labels[np.argmax(probas[0])] == 'woman':
            female += 1
            
    n_male.append(male)
    n_female.append(female)

    
df_frames_1["n_faces"] = n_faces
df_frames_1["n_male"] = n_male
df_frames_1["n_female"] = n_female
```

### Conclusion and interpretation

To answer our research question, we have to look at the proportions of detected males and females. For the first time period 42.9% of the detected faces was classified as female. For the second time period 32.3% of the detected faces was classified as female. For the third time period 41.7% percent of the detected faces was classified as female. According to these results, we can answer our research question with "no, we did not find that the proportion of females increased over our chosen time periods". 

But, there are some things which could have been the cause of this result. The first time period had only selected 21 faces. This number is very low and could be the cause of this high number. This low amount of detected faces could be the cause of poor video quality. Also the second timeperiod has selected a quite small amount of faces. Another explanation for our results could have been the choice for frames per scene. Afterall choosing a method to extract one frame per second could have led to better results.

So, we conclude that our results are likely to be highly inaccurate. Follow-up research should use another way to select the frames. Hopefully this well lead to a higher amount of detected faces with as result a better detection of the proportion of female actresses.
